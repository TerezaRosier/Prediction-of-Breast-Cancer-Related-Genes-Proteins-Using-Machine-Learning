#renaming the synomym GeneIDs as per official names
#both Harmozinome and DisGENet utilize the "NCBI Entrez" for gene identification. 

```python
def rename_genes_in_dataframe(df, gene_input):
    """
    Function to rename gene synonyms in a DataFrame.

    Parameters:
    df (DataFrame): The DataFrame where the genes will be renamed.
    gene_input (str): String with gene synonyms and common name in the format 'synonyms; common_name'.

    Returns:
    DataFrame: The updated DataFrame with genes renamed.
    """
    # Separate the synonyms from the common name
    gene_synonyms, common_name = gene_input.split(';')
    
    # Convert the synonym string into a list
    synonym_list = [gene.strip() for gene in gene_synonyms.split(',')]
    
    # Replace the synonyms with the common name in the DataFrame
    df.replace(synonym_list, common_name.strip(), inplace=True)
    
    return df
```

```python
# List of gene inputs to be renamed
gene_inputs = [
    "BCC7, LFS1, P53, TRP53; TP53",
    "BRCAI, BRCC1, BROVCA1, FANCS, IRIS, PNCA4, PPP1R53, PSCP, RNF53; BRCA1",
    "AT1, ATA, ATC, ATD, ATDC, ATE, TEL1, TELO1; ATM",
    "BRCC2, BROVCA2, FACD, FAD, FAD1, FANCD, FANCD1, GLM3, PNCA2, XRCC11, RP11-298P3.4; BRCA2",
    "10q23del, BZS, CWS1, DEC, GLM2, MHAM, MMAC1, PTEN1, TEP1; PTEN",
    "CDS1, CHK2, HuCds1, LFS2, PP1425, RAD53, hCds1, RP11-436C9.1; CHEK2",
    "CLOVE, CWS5, MCAP, MCM, MCMTC, PI3K, p110-alpha; PIK3CA"]

# Loop to apply the function to each entry
for gene_input in gene_inputs:
    df = rename_genes_in_dataframe(df, gene_input)
```

```python
# List of genes to keep
genes_to_keep = ['TP53', 'BRCA1','ATM', 'BRCA2', 'PTEN', 'CHEK2', 'PIK3CA']

# Filtering the DataFrame to keep only the specific genes
df = df[df['Gene/Protein Symbol'].isin(genes_to_keep)].reset_index(drop=True)

# Displaying the updated DataFrame
df
```
